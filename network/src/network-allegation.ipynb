```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def load_data():
    dfa = pd.read_csv("../data/input/event.csv")

    dfb = pd.read_csv("../data/input/allegation.csv")

    dfb = dfb[["allegation_uid", "supervisor_uid", "tracking_id", "assigned_department", "assigned_division", "investigator_uid"]]

    df = pd.merge(dfa, dfb, on="allegation_uid", how="outer")
    return df

events = load_data()
```

```python
def load_index():
    df = pd.read_csv("../../preprocess/data/output/clean.csv")
    return df 

wrongful_conviction_index = load_index()
wrongful_conviction_index
event_table = events.copy()

wrongful_conviction_merge = pd.merge(wrongful_conviction_index, event_table, on="uid")

wrongful_conviction_merge = wrongful_conviction_merge[~((wrongful_conviction_merge.tracking_id.fillna("") == ""))]
```

```python
merged_table = pd.merge(event_table, wrongful_conviction_merge[["uid", "officer_name"]], on='uid', how='left')

merged_table = merged_table[~((merged_table.tracking_id.isnull()))]
merged_table = merged_table[~((merged_table.tracking_id.fillna("") == ""))]

connections = {}
for _, row in merged_table.iterrows():
    officer_name = row['officer_name']
    tracking_id = row['tracking_id']
    
    if pd.isnull(officer_name) or pd.isnull(tracking_id):
        continue  # Skip the row if either officer_name or tracking_id is missing or NaN
    
    if officer_name in wrongful_conviction_merge['officer_name'].values:
        # Find all unique officers from the event table who share the same tracking_id
        related_officers = merged_table[(merged_table['tracking_id'] == tracking_id) & (merged_table['officer_name'] != officer_name)]['officer_name'].unique()
        if len(related_officers) > 0:
            connections[officer_name] = {'related_officers': related_officers, 'tracking_id': tracking_id}

G = nx.Graph()
for officer, data in connections.items():
    if not pd.isnull(officer):
        G.add_node(officer, type='Wrongful Conviction Officer')
        for related_officer in data['related_officers']:
            if not pd.isnull(related_officer):
                G.add_node(related_officer, type='Related Officer')
                G.add_edge(officer, related_officer, tracking_id=data['tracking_id'])

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, iterations=50)
node_colors = ['lightblue' if G.nodes[node]['type'] == 'Wrongful Conviction Officer' else 'lightgreen' for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=200, node_color=node_colors)
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.7)
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10)
plt.axis('off')
plt.title('Officer Connections based on Shared Tracking ID')
plt.tight_layout()
plt.show()

output_table = pd.DataFrame(columns=['Wrongful Conviction Officer', 'Related Officer', 'Shared Tracking ID'])
for officer, data in connections.items():
    if not pd.isnull(officer):
        for related_officer in data['related_officers']:
            if not pd.isnull(related_officer):
                new_row = pd.DataFrame({
                    'Wrongful Conviction Officer': [officer],
                    'Related Officer': [related_officer],
                    'Shared Tracking ID': [data['tracking_id']]
                })
                output_table = pd.concat([output_table, new_row], ignore_index=True)

output_table.to_csv('officer_tracking_id_connections.csv', index=False)
```
